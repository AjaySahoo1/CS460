```python
import sklearn.datasets
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split



url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
# download and convert the csv into a DataFrame
df = pd.read_csv(url, header=None)
df.head()
```

```python
# extract the label column
z = df.iloc[:, 4].values
# extract features
x = df.iloc[:, 0:1].values
y = df.iloc[:, 1:2].values

x = x[0:100]
y=y[0:100]
z = z[0:100]
print(x)
print (y)
for i in range(z.shape[0]):
    if(z[i]=='Iris-setosa'):
        z[i]=1
    else:
        z[i]=-1
print(z)
# plot Iris Setosa samples
plt.scatter(x[:50], y[:50], color='red', marker='o', label='Setosa')
# plot Iris Versicolour samples
plt.scatter(x[50:100], y[50:100], color='blue', marker='x', label='Versicolour')

# show the legend
plt.xlabel("Sepal length")
plt.ylabel("Petal length")
plt.legend(loc='upper left')

# show the plot
plt.show()
```

```python
# split the data
x_train, x_test, y_train, y_test, z_train, z_test = train_test_split(x, y,z, test_size=0.25,
                                                    random_state=0)
print(x_train)
print(y_train)
print(z_train)
```

```python

def Perceptron(x,y,z,iter):
    w=np.zeros(len(x))

    b=0

    threshold=0
    #iter= 100

    for k in range(iter):
        total=0
        for i in range(0,len(x)):
            sum=0
            for j in range(0,len(x[i])-1):
                sum+= x[j]*w[j]
            output= b+sum
        
            if output>threshold:
                a=1
            else:
                a=-1
            if a*z[i]<=0:
                for m in range(0,len(w)):
                    w[m]+=y[i]*x[m]
                b+=y[i]
                total+=1
    return [w,b,total]

thita1= Perceptron(x_train,y_train,z_train,100)

print(thita1)
```

```python
def avgPerceptron(x,y,z,iter):
    w1=np.zeros(len(x))

    b1=0
    w2=np.zeros(len(x))

    b2=0
    c=1
    threshold=0
    iter= 100

    for k in range(iter):
        total2=0
        for i in range(0,len(x)):
            sum=0
            for j in range(0,len(x[i])-1):
                sum+= x[j]*w1
            output= b1+sum
        
            if output>threshold:
                a=1
            else:
                a=-1
            if a*z[i]<=0:
                for m in range(0,len(w1)):
                    w1[m]+=y[i]*x[m]
                for n in range(0,len(w2)):
                    w2[n]+=y[i]*x[n]*c
                b1+=y[i]
                b2+=y[i]*c
                total2+=1
            c+=1
            c=int(c)
    for t in range(0,len(w1)):
                w1[t]=w1[t]-w2[t]/c
    return [w1,b1-b2*(1/c),total2]

thita2= avgPerceptron(x_train,y_train,z_train,100)

print(thita2)
```
